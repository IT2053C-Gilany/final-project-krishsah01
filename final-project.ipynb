```python
import json
import zipfile

with zipfile.ZipFile('data/pollution-by-electric-vehicles-dataset.json.zip', 'r') as zip_ref:
  zip_ref.extractall('data/')

with open('data/pollution-by-electric-vehicles-dataset.json', 'r') as f:
  data = json.load(f)

print(f"Data type: {type(data)}")
if isinstance(data, dict):
  print(f"Keys in the data: {list(data.keys())}")

if isinstance(data, dict) and 'data' in data:
  df = pd.DataFrame(data['data'])
  print(f"\nDataset shape: {df.shape}")
  print("Successfully loaded dataset from 'data' key")
else:
  print("Could not find 'data' key in JSON structure")

print("\nFirst 3 rows:")
df.head(3)
```
